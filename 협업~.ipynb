## 협업필터링 적용

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# # 임의의 사용자-책 평점 데이터
# data = {
#     'User': ['User1', 'User1', 'User2', 'User2', 'User3', 'User3', 'User4', 'User4'],
#     'Book': ['Book1', 'Book2', 'Book2', 'Book3', 'Book1', 'Book3', 'Book2', 'Book4'],
#     'Rating': [5, 4, 3, 2, 4, 5, 4, 5]
# }

# df = pd.DataFrame(data)

# Pivot table을 사용하여 사용자-책 평점 매트릭스 생성
user_book_rating = collaborative_df.pivot_table(index='user_id', columns='isbn_id', values='rating').fillna(0)

# 코사인 유사도 계산
user_similarity = cosine_similarity(user_book_rating)
user_similarity_df = pd.DataFrame(user_similarity, index=user_book_rating.index, columns=user_book_rating.index)

# 사용자 기반 추천
def get_user_recommendations(target_user):
    # 타겟 사용자와 다른 사용자 간의 유사도
    similar_users = user_similarity_df[target_user]
    
    # 유사도 순으로 정렬
    similar_users = similar_users.sort_values(ascending=False)
    
    # 가장 유사한 사용자
    most_similar_user = similar_users.index[1]
    
    # 가장 유사한 사용자의 평점을 가져옴
    user2_ratings = user_book_rating.loc[most_similar_user, :]
    
    # 타겟 사용자의 평점을 가져옴
    user1_ratings = user_book_rating.loc[target_user, :]
    
    # 타겟 사용자가 평가하지 않은 책만 추천
    recommendations = user2_ratings[user1_ratings == 0]
    
    # 평점 순으로 정렬하여 추천
    recommendations = recommendations.sort_values(ascending=False)
    
    return recommendations

print(get_user_recommendations(4))
```

### 일단 책 정보 데이터(data.pkl)에서 isbn_id 빼오고, user_id 20명으로, rating 0~5로 랜덤 부여해서 데이터 생성함
### rating 0이면 평가 안한 것. 이 데이터로 협업 필터링 기초적인 동작은 하도록 완료!
