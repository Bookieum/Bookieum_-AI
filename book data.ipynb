```python
import nltk
nltk.download('stopwords')
from mecab import MeCab
mecab = MeCab()
```

```python
def extract_nouns(text, target = ['NNG','NNP'], to_list=True):
    # ['NNG','NNP','NP','NNB','NNBC','NR']

    nouns_list = []
    pod_text = mecab.pos(text)
    for s in pod_text:
        if s[1] in target:
            if to_list:
                nouns_list.append(s[0])
            else:
                nouns_list.append(s)

    return nouns_list
```

## 도서별 키워드

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# TfidfVectorizer 객체를 저장
cv_list = []

# Term-Document Matrix(TDM)를 저장
# TDM은 문서와 단어의 빈도를 표현한 행렬
tdm_list = []

# 각 텍스트에서 선택된 단어들을 저장, 30개씩
words_list = []

#  TDM에서 각 열(단어)에 대해 계산한 합계를 저장
# 당 단어가 전체 텍스트에서 몇 번 등장했는지
count_mat_list = []

#  'count_mat'을 1차원 배열로 변환하여 저장
count_list = []
```

```python
for i, sublist in enumerate(nouns_list):
    if not sublist:  # 서브리스트가 비어 있는지 확인
        print(f"Sublist at index {i} is empty.")
    else:
        for j, item in enumerate(sublist):
            if not item:  # 항목이 비어 있는지 확인
                print(f"Item at index {i},{j} is empty.")
```

```python
for i in range(len(nouns_list)):
    tfidf_vectorizer = TfidfVectorizer(max_features=30) # feature 개수 임의 설정

    tdm = tfidf_vectorizer.fit_transform(nouns_list[i])
    words = tfidf_vectorizer.get_feature_names_out()
    words_list.append([words])

    count_mat = tdm.sum(axis=0)  # 단어별로 count 매김
    count = np.squeeze(np.asarray(count_mat))  # squeeze : 차원을 하나 줄임 (2차원 >> 1차원)

    count_list.append([count])
```

```python
keyword = pd.DataFrame(words_list)
df1['Keyword'] = keyword
df1.head()
```

```python
import operator
```

```python
word_count_list = []
for word, count in zip(words_list, count_list):
    word_count = list(zip(word, count))[0]
    #word_count = sorted(word_count, key=operator.itemgetter(1), reverse=True)
    word_count_list.append(word_count)
```
