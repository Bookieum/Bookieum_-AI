```python
def extract_keywords(text):
    # 형태소 분석을 통해 명사만 추출
    nouns = okt.nouns(text)

    return ' '.join(nouns)

# 'keywords'라는 새로운 컬럼에 명사만 추출한 결과 저장
ex3['keywords'] = ex3['BOOK_INTRCN_CN'].apply(extract_keywords)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TfidfVectorizer 인스턴스 생성
tfidf = TfidfVectorizer()
tfidf_title = TfidfVectorizer()

# 'keywords' 컬럼에 대해 TF-IDF 벡터화 수행
tfidf_matrix = tfidf.fit_transform(ex3['keywords'])
tfidf_matrix_title = tfidf_title.fit_transform(ex3['TITLE_NM'])

print(tfidf_matrix.shape, tfidf_matrix_title.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 코사인 유사도 계산
cosine_sim = cosine_similarity(tfidf_matrix)

print(cosine_sim)

```

```python
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

def recommend_books_based_on_sentence(sentence, tfidf, tfidf_matrix, data):
    # KoNLPy의 Okt 형태소 분석기 객체 생성
    okt = Okt()

    # 사용자가 입력한 문장 (형태소로 분리)
    user_input_sentence = [" ".join(okt.morphs(sentence))]

    # 이 문장도 TF-IDF 벡터화를 수행합니다.
    user_input_tfidf = tfidf.transform(user_input_sentence)

    # 코사인 유사도 계산
    cosine_sim_user_input = cosine_similarity(user_input_tfidf, tfidf_matrix)

    # 가장 유사한 10개의 문서 인덱스 찾기
    top_10_indexes = np.argsort(-cosine_sim_user_input.flatten())[:10]

    # 원본 데이터프레임에서 해당하는 행 가져오기 (여기서는 'TITLE_NM' 컬럼만 가져옵니다.)
    similar_books_based_on_user_input = data['TITLE_NM'].iloc[top_10_indexes]

    return similar_books_based_on_user_input

```

```python
sentence = "인간의 삶은 종교적 신념에 크게 영향을 받는다. 우리의 이해와 가치 판단, 그리고 행동에까지 미치는 종교의 역할은 매우 중요하다. 어떤 사람들은 그것을 통해 의미를 찾고, 다른 사람들은 위안을 얻거나 공동체 의식을 공유한다. 우리 모두 서로 다른 신념과 경험을 존중하며 평화롭게 공존해 나가자."
```

```python
# 사용 예시:

recommendations = recommend_books_based_on_sentence(sentence, tfidf, tfidf_matrix, ex3)
print(recommendations)
```

```python
# 각 책에 대해 가장 유사한 다른 10권의 책 찾기
indices = pd.Series(ex.index, index=ex['TITLE_NM']).drop_duplicates()
def get_recommendations(title_nm, cosine_sim=cosine_sim):
    idx = indices[title_nm]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    book_indices = [i[0] for i in sim_scores]
    return ex['TITLE_NM'].iloc[book_indices]
```

```python
print(get_recommendations('영유아 행동지도'))
```
